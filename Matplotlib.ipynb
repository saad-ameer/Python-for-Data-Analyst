## Types of Data

When deciding how to visualize your data, it’s important to understand the **type of data** you're working with.

### 1. **Categorical Data (Dimensions)**
- Example: Office Location  
- Also called: Qualitative data, non-contiguous data  
- Represents: Names, categories, or fixed labels  
- Key trait: Finite set of values (even if large)  

### 2. **Numerical Data (Measures)**
- Examples: Revenue, Units Sold  
- Also called: Quantitative data, continuous data  
- Represents: Numeric values that can be aggregated  
- Key trait: Can theoretically take infinite values  

---

|------------------|-----------------------------------------------------------|---------------------------|
| **Bar Chart**     | Compare values across categories                          | 1 Dimension + 1 Measure   |
| **Scatter Plot**  | Explore relationship or correlation between variables     | 2 Measures                |

### Examples:

- To compare **revenue across office locations**, use a **bar chart**.  
- To analyze the **relationship between revenue and units sold**, use a **scatter plot**.

# Creating a Plotting Area in Matplotlib (Object-Oriented Approach)

Before plotting any data, it’s important to set up the **plotting area**. This includes configuring:
- The figure size
- Axes positions and dimensions
- X and Y axis range
- Titles and labels

Matplotlib uses an **object-oriented approach** to plotting, which allows you to build and customize plots step by step.

---

## Step 1: Import Required Libraries

```python
import matplotlib.pyplot as plt
```

---

## Step 2: Create a Figure Object

The `figure()` method creates a figure object:

```python
fig = plt.figure()
```

- By default, the figure size is 6.4 x 4.8 inches.
- You can customize it using the `figsize` parameter.

```python
fig = plt.figure(figsize=(10, 10))  # 10x10 inches
```

---

## Step 3: Add Axes

You can add axes using the `add_axes()` method. It takes a list of four values:

```python
ax = fig.add_axes([x_pos, y_pos, width, height])
```

All values are fractions of the figure's size.

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_axes([0, 0, 1, 1])  # Full plot area
```

### Examples

```python
# Half width and height
ax = fig.add_axes([0, 0, 0.5, 0.5])

# Full width, half height
ax = fig.add_axes([0, 0, 1, 0.5])
```

---

## Step 4: Organize Axes with Variables

A cleaner approach is to store your axes in separate variables:

```python
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([1.5, 0, 1, 1])  # Moved to the right
```

- `x_pos` and `y_pos`: position from left and bottom
- `width` and `height`: size as fraction of the figure

---

## Step 5: Move Axes and Nest Them

```python
# Nested axes
ax2 = fig.add_axes([0.1, 0.1, 0.5, 0.5])
```

---

## Step 6: Set Axis Limits

```python
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 100)

ax2.set_xlim(0, 50)
ax2.set_ylim(0, 50)
```

---

## Step 7: Twin Axes (Dual Y-Axis)

You can create two y-axes on the same plot using `twinx()`:

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = ax1.twinx()

ax1.set_ylim(0, 100)
ax2.set_ylim(0, 200)
```

---

## Step 8: Add Titles and Labels

```python
ax1.set_title("Chart 1")
ax2.set_title("Chart 2")

ax1.set_xlabel("X-Axis")
ax1.set_ylabel("Y-Axis")

ax2.set_xlabel("X-Axis")
ax2.set_ylabel("Y-Axis")
```

---

## Summary

- Use `plt.figure()` to create a canvas
- Use `add_axes()` to place plotting areas
- Use `set_xlim()`, `set_ylim()` to define value ranges
- Use `set_title()`, `set_xlabel()`, `set_ylabel()` to label plots
- Use `twinx()` for dual axes

# Creating Subplots with Matplotlib

The `subplots()` method is a convenient way to create multiple plots. It automatically generates both:
- A **figure** object
- One or more **axes** objects

This is much simpler and more efficient than manually using `add_axes()` multiple times.

---

## Syntax Overview

```python
fig, ax = plt.subplots(nrows, ncols)
```

- `nrows`: Number of rows of subplots
- `ncols`: Number of columns of subplots
- Returns:
  - `fig`: Figure object
  - `ax`: Single Axes object (if `nrows*ncols == 1`) or
  - Numpy array of Axes objects (if more than one subplot)

---

## Basic Examples

### Single Plot (1 Row, 1 Column)

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
print(type(fig))  # <class 'matplotlib.figure.Figure'>
print(type(ax))   # <class 'matplotlib.axes._subplots.AxesSubplot'>
```

---

### Horizontal Subplots (1 Row, 2 Columns)

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.set_title("Left Plot")
ax2.set_title("Right Plot")
```

You can also assign and access like this:

```python
fig, ax_arr = plt.subplots(1, 2)
ax_arr[0].set_title("Left Plot")
ax_arr[1].set_title("Right Plot")
```

---

### 2x2 Subplots (2 Rows, 2 Columns)

```python
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2, 2)
ax1.set_title("Top Left")
ax2.set_title("Top Right")
ax3.set_title("Bottom Left")
ax4.set_title("Bottom Right")
```

---

## Improving Layout

Sometimes subplots may overlap. To fix this, set:

```python
plt.subplots(nrows, ncols, tight_layout=True)
```

This adjusts spacing to prevent overlaps.

---

## Summary

| Pattern              | Shape of `ax`                | Access Pattern       |
|----------------------|------------------------------|----------------------|
| `1 x 1`              | Single Axes object            | `ax`                 |
| `1 x N` or `N x 1`   | 1D NumPy array of Axes        | `ax[0]`, `ax[1]`, …  |
| `M x N` (M,N > 1)    | 2D NumPy array of Axes        | `ax[0][1]`, `ax[1][0]`, etc. |

# Bar Plot with Matplotlib

* Bar plots are used to visualize **categorical data**, where the height/width of each bar represents the value of that category.
* Vertical bar plot: use `ax.bar(x, height)`
* Horizontal bar plot: use `ax.barh(y, width)`

## Data Preparation

* Use `pandas` to load and manipulate the dataset.
* Group by `region` and calculate total `population` using `.groupby()` and `.sum()`.
* Use `.reset_index()` so that `region` becomes a column again.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("top10_countries.csv")
df_grouped = df[['region', 'population']].groupby('region').sum().reset_index()
```

## Creating a Basic Vertical Bar Plot

* Use `plt.figure()` to create a figure and `fig.add_axes()` to create a custom axes object.
* Coordinates `[left, bottom, width, height]` are in fractions of the figure.

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlabel("Region")
ax.set_ylabel("Total Population")
ax.set_title("Population by Region")
ax.bar(df_grouped["region"], df_grouped["population"])
plt.show()
```

## Customizing Bars

* Use `color='blue'` or any valid string, letter (e.g. `'r'`), or hex code for fill color.
* Use `edgecolor='black'` to change the border color.
* To apply different colors for each bar, pass a list of color codes.

```python
colors = ['b', 'r', 'y', 'g']
ax.bar(df_grouped["region"], df_grouped["population"], color=colors, edgecolor='black')
```

## Sorting Bars

* Use `.sort_values(by='population', ascending=False)` to sort bars in descending order of height.

```python
df_grouped.sort_values(by='population', ascending=False, inplace=True)
```

## Common Mistake – Not Grouping Data

* If you plot raw country-level data without grouping, you'll get multiple bars for the same region.
* Bars won't stack and totals will be incorrect.
* Always group and aggregate your data before plotting group totals.

## Horizontal Bar Plot

* Use `.barh()` to plot horizontal bars.
* `y` is the category and `width` is the value.

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Population by Country")
ax.set_xlabel("Population")
ax.set_ylabel("Country")
ax.barh(df["country"], df["population"], color='white', edgecolor='red')
plt.show()
```

## Side-by-Side Vertical and Horizontal Plots

* Use `plt.subplots(1, 2)` to create a figure with two axes objects.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Vertical Bar Plot
ax1.set_title("Population by Region")
ax1.set_xlabel("Region")
ax1.set_ylabel("Population")
ax1.bar(df_grouped["region"], df_grouped["population"], color='blue', edgecolor='black')

# Horizontal Bar Plot
ax2.set_title("Population by Country")
ax2.set_xlabel("Population")
ax2.set_ylabel("Country")
ax2.barh(df["country"], df["population"], color='white', edgecolor='black')

plt.tight_layout()
plt.show()
```

## Summary

* Use `.bar()` for vertical and `.barh()` for horizontal bar plots.
* Always prepare your data with proper grouping and aggregation.
* Customize bars with `color`, `edgecolor`, and `figsize`.
* Use `plt.subplots()` for multiple side-by-side visualizations.

# Line Plot with Matplotlib

* Line plots are ideal for visualizing trends over time (e.g. date or year-wise data).
* Matplotlib doesn’t have a dedicated `line()` function, but the `plot()` method is used to draw line charts.
* The first argument is for x-axis (e.g. year), and the second and onward are for y-values (measures).

## Data Preparation

* Use `pandas` and `matplotlib.pyplot` for analysis and plotting.
* Convert the `date` column to datetime using `pd.to_datetime()`.
* Extract the year using `dt.strftime('%Y')`.
* Group data using `.groupby('year')` and get the yearly sum of the target column (e.g. `bikescot`).
* Use `.reset_index()` to make `year` a regular column.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("AFL_DailyBikeCase.csv")
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")
df["year"] = df["date"].dt.strftime("%Y")

df_yearly = df.groupby("year")["bikescot"].sum().reset_index()
```

## Basic Line Plot

* `plot(x, y)` requires both arguments to be **positional**, not keyword.
* You can also use the `data` parameter to pass a dataframe and use column names directly.

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(df_yearly["year"], df_yearly["bikescot"])
plt.show()
```

Or using `data`:

```python
ax.plot("year", "bikescot", data=df_yearly)
```

## Customizing Line Style

* Use `color` for line color (`'blue'`, `'red'`, `'#123456'`, etc.).
* Use `alpha` for transparency (`0.0` = fully transparent, `1.0` = opaque).
* Use `marker` to add symbols at data points (e.g. `'o'`, `'s'`, `'1'`).
* Use `markersize` to control marker size.
* Use `linestyle` for line style (e.g. `':'`, `'--'`, `'-.'`).

```python
ax.plot("year", "bikescot", data=df_yearly, color='red', alpha=0.5, marker='o', markersize=8, linestyle=':')
```

## Plotting Subset of Data

* Use `.tail()` to plot only recent years, such as the last 10 years.

```python
recent = df_yearly.tail(10)
ax.plot("year", "bikescot", data=recent)
```

## Plotting Multiple Measures (Optional)

* You can pass multiple y-values as positional arguments to compare trends on the same x-axis.

```python
# Example with synthetic column for demonstration
ax.plot(df_yearly["year"], df_yearly["bikescot"], df_yearly["bikescot"] * 2)
```

## Summary

* Use `plot(x, y)` for line charts; both x and y must be **positional arguments**.
* Customize visuals with `color`, `alpha`, `marker`, `linestyle`, etc.
* Use `.tail()` to slice time series for recent trends.
* Always group and aggregate before plotting when working with time-based data.

# Scatter Plot with Matplotlib

* Scatter plots are used to visualize the **relationship between two numerical variables**.
* They are helpful for identifying **correlations, outliers**, and **clusters**.
* Each point represents a record, with optional size variation to add another dimension.

## Dataset Overview

* FIFA 21 player dataset contains attributes like `age`, `overall` rating, `wage`, and `value`.
* Load only the top 100 players for focused visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt

players = pd.read_csv("players_21.csv")
players_100 = players.head(100)
```

## Basic Scatter Plot

* Use `ax.scatter()` for creating scatter plots.
* You can pass `x`, `y`, and `data` arguments to reference columns by name.

```python
fig = plt.figure(figsize=(6, 5))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlabel("Overall Rating")
ax.set_ylabel("Wage in Euros")

ax.scatter(x="overall", y="wage_eur", data=players_100)
plt.show()
```

## Adjusting Point Size

* Use `s` argument to set the size of markers.
* Pass a **scalar** value to resize all points equally.
* Pass a **column** (e.g. `value_eur`) to vary point sizes based on a third variable.
* Normalize large values to make plots readable.

```python
# Fixed size
ax.scatter(x="overall", y="wage_eur", data=players_100, s=100)

# Variable size based on player value
ax.scatter(x="overall", y="wage_eur", data=players_100, s=players_100["value_eur"] / 2)
```

## Summary

* Use `scatter(x, y, data=..., s=...)` to visualize relationships between measures.
* Set `s` to add size-based dimension (e.g. player value).
* Normalize `s` if values are too large to make bubbles visible.
* Scatter plots are effective for visualizing patterns, but color-based categories are better handled with Seaborn.

# Box Plot and Violin Plot with Matplotlib

## Overview

* **Box Plot** visualizes the distribution of data using quartiles and outliers.
  * Shows min, Q1, median (Q2), Q3, max (whiskers), and outliers.
* **Violin Plot** adds a kernel density estimation to show data distribution shape.

## Dataset

* Using **FIFA 21 player ratings** dataset.
* Analyzing the `overall` score of all players and comparing scores between two clubs.

```python
import pandas as pd
import matplotlib.pyplot as plt

players = pd.read_csv("players_21.csv")
```

## Basic Box Plot

* Use `ax.boxplot()` and pass a list of numeric arrays.
* Use `labels` to name each box.
* `showfliers=False` removes outliers (shown as circles).

```python
fig = plt.figure(figsize=(5, 4))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Overall Rating Distribution")

# Overall scores box plot
ax.boxplot([players["overall"]], labels=["Overall Score"], showfliers=False)
plt.show()
```

## Box Plot for Two Teams

* Filter `overall` scores for specific clubs: Inter and Dortmund.
* Pass both arrays as a list to `boxplot()`.

```python
inter = players[players["club_name"] == "Inter"]["overall"]
dortmund = players[players["club_name"] == "Borussia Dortmund"]["overall"]
data = [inter, dortmund]

fig = plt.figure(figsize=(6, 4))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Team-wise Overall Ratings")

ax.boxplot(data, labels=["Inter", "Dortmund"], showfliers=False)
plt.show()
```

## Violin Plot (Using `ax.violinplot()`)

* Similar to box plot but shows frequency density as width.
* `positions` define the x-axis position for each violin.

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Violin Plot of Team Ratings")

ax.violinplot(data, positions=[1, 2])
ax.set_xticks([1, 2])
ax.set_xticklabels(["Inter", "Dortmund"])
plt.show()
```

## Combined Box and Violin Plot

* Combine `violinplot()` and `boxplot()` on same axes for richer visualization.

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Combined Violin and Box Plot")

ax.violinplot(data, positions=[1, 2])
ax.boxplot(data, positions=[1, 2], widths=0.2, labels=["Inter", "Dortmund"], showfliers=False)
plt.show()
```

## Summary

* Use `boxplot()` to visualize quartiles and outliers.
* Use `violinplot()` to view density and shape of distribution.
* Combine both for deeper insight.
* Great for comparing distributions across groups.

# Matplotlib Formatting & Styling – Crisp Notes

## Overview
* This guide consolidates key formatting and styling tips in `matplotlib`.
* It helps you create clean, informative, and visually appealing plots.

---

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 100, 0.1)
y1 = x ** 2
y2 = x ** 3
```

---

## 1. Layout Adjustment

* Use `tight_layout()` to fix overlap in subplots.
```python
fig, axs = plt.subplots(2, 2)
plt.tight_layout()
```

---

## 2. Labels, Titles & Fonts

* Use `fontsize`, `fontname`, `color`, and `loc` for custom titles and labels.
* Rotate labels using `rotation`.
```python
fig, ax = plt.subplots()
ax.plot(x, y1)
ax.set_title("My Chart", fontsize=14, fontname='Arial', color='blue', loc='left')
ax.set_xlabel("X Axis", fontsize=12, rotation=0)
ax.set_ylabel("Y Axis", fontsize=12, color='green')
```

---

## 3. Tick Customization

* Customize ticks using `set_xticks()` and `set_xticklabels()`.
* Rotate ticks for better readability.
```python
ax.set_xticks([0, 20, 40, 60])
ax.set_xticklabels(["Zero", "Twenty", "Forty", "Sixty"], rotation="vertical")
```

---

## 4. Legends

* Add `label` in `plot()` then call `legend()`.
* Control legend position with `loc`.
```python
ax.plot(x, y1, label='x^2')
ax.plot(x, y2, label='x^3')
ax.legend(loc='upper left')
```

---

## 5. Legends for Bar Plots

* When plotting bars in loops, assign a label per bar.
```python
for i in range(len(df)):
    ax.bar(df["country"][i], df["population"][i], label=df["country"][i])
ax.legend()
```

---

## 6. Annotations

* Use `annotate()` to add text with optional arrows.
```python
ax.annotate("Important", xy=(50, 2500), xytext=(60, 5000),
            arrowprops=dict(facecolor='black', shrink=0.05))
```

---

## 7. Colors

* Use named colors (`"red"`), short codes (`'r'`), or hex codes (`"#1f77b4"`).
```python
ax.plot(x, y1, color='royalblue')
```

---

## 8. Gridlines

* Use `grid()` to add and style grids.
```python
ax.grid(True, linestyle='--', color='gray', axis='both')
```

---

## 9. Stylesheets

* Use `plt.style.use()` to apply predefined themes like `"ggplot"` or `"dark_background"`.
```python
plt.style.use("seaborn")
```

---

* `tight_layout()` removes subplot clutter.

* `legend()` uses plot `label=` tags.
* `annotate()` adds informative text/arrows.
* Use color names, hex, or short codes.
